# Lab02 - See the World through Convolution

```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
import numpy as np
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
from torch import nn
```

## Why Convolution?



```python
def plot_convolution(image, kernel):
    image_tensor = torch.tensor(image, dtype=torch.float32).unsqueeze(0).unsqueeze(0)
    kernel_tensor = torch.tensor(kernel, dtype=torch.float32).unsqueeze(0).unsqueeze(0)
    with torch.no_grad():
        convolved = F.conv2d(image_tensor, kernel_tensor, padding=1)
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 3, 1)
    plt.title("Original")
    plt.imshow(image, cmap='gray')
    plt.subplot(1, 3, 2)
    plt.title("Kernel")
    plt.imshow(kernel, cmap='gray')
    plt.subplot(1, 3, 3)
    plt.title("After Convolution")
    plt.imshow(convolved.squeeze().numpy(), cmap='gray')
    plt.show()

# Example: edge detection
image = np.array([[0, 0, 0, 0, 0],
                  [0, 255, 255, 255, 0],
                  [0, 255, 0, 255, 0],
                  [0, 255, 255, 255, 0],
                  [0, 0, 0, 0, 0]], dtype=np.float32)

sobel_x = np.array([[1, 0, -1],
                    [2, 0, -2],
                    [1, 0, -1]], dtype=np.float32)

plot_convolution(image, sobel_x)
```
